```python
import subprocess

# Run the whisper command with --word_timestamps True and capture its output
command = "whisper 'audio.mp3' --model large"
process = subprocess.run(command, shell=True, check=True, text=True, capture_output=True)

# The stdout attribute contains the output of the command
output = process.stdout
```

```python
# Now output contains the transcription and timestampsprint(output)
print(output)
```

```python
# Split the output into lines
lines = output.split('\n')

# Initialize empty lists to store the timestamps and text
timestamps = []
texts = []

# Loop over each line
for line in lines:
    # Check if the line contains the '] ' delimiter
    if '] ' in line:
        # Split the line into timestamp and text
        timestamp, text = line.split('] ', 1)

        # Remove the leading '[' from the timestamp
        timestamp = timestamp[1:]

        # Append the timestamp and text to their respective lists
        timestamps.append(timestamp)
        texts.append(text)

# Now timestamps and texts are separate lists
print("Timestamps:", timestamps)
print("Texts:", texts)
```

```python
# Join the texts into a single string
text_string = ' '.join(texts)

# Now text_string contains all the text as a single string
print("Text String:", text_string)
```

The Below code perform speaker segmentation but this one saves the audios

```python
from pyannote.audio import Pipeline
from pydub import AudioSegment
import torch

pipeline = Pipeline.from_pretrained(
    "pyannote/speaker-diarization-3.1",
    use_auth_token="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")

# send pipeline to GPU (when available)
pipeline.to(torch.device("cuda"))

# apply pretrained pipeline
diarization = pipeline("audio.mp3")

# load audio file
audio = AudioSegment.from_mp3("audio.mp3")

# print the result and save audio chunks
for turn, _, speaker in diarization.itertracks(yield_label=True):
    print(f"start={turn.start:.2f}s stop={turn.end:.2f}s speaker_{speaker}")

    # convert start and end times to milliseconds
    start_time = int(turn.start * 1000)
    end_time = int(turn.end * 1000)

    # extract chunk
    chunk = audio[start_time:end_time]

    # save chunk to a new file
    chunk.export(f"speaker_{speaker}_{start_time}_{end_time}.mp3", format="mp3")
```
