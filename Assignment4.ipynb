```python
# Import important libraries
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
from scipy.spatial.distance import cityblock
import pandas as pd
import math
```

```python
docs = [
    "MAchine learning is on of the prominant domain of computer science",
    "this is one of the best machine learning courses",
    "machine learning Engineer perform machine learning techniques"
]
```

```python
# Creating a Bag-of-Words model
bow_vectorizer = CountVectorizer()
bow_matrix = bow_vectorizer.fit_transform(docs)
bow_tokens = bow_vectorizer.get_feature_names_out()
bow_df = pd.DataFrame(data=bow_matrix.toarray(), columns=bow_tokens)
print("\nBag-of-Words Matrix:")
print(bow_df)
```

```python
# Calculating Term Frequency
def calc_tf(doc_list):
    tf_list = []
    for doc in doc_list:
        tf_dict = {}
        total_words = len(doc.split())
        for word in doc.split():
            tf_dict[word] = tf_dict.get(word, 0) + 1
            tf_dict[word] = round(tf_dict[word] / total_words, 2)
        tf_list.append(tf_dict)
    return tf_list

tf_results = calc_tf(docs)
tf_df = pd.DataFrame(tf_results).fillna(0)
print("\nTerm Frequency Matrix:")
print(tf_df)

# Calculating Inverse Document Frequency
def calc_idf(doc_list):
    N = len(doc_list)
    term_doc_count = {}
    for doc in doc_list:
        unique_terms = set(doc.split())
        for term in unique_terms:
            term_doc_count[term] = term_doc_count.get(term, 0) + 1
    idf_dict = {}
    for term, count in term_doc_count.items():
        idf_dict[term] = round(math.log(N / (1 + count)), 2)
    return idf_dict

idf_dict = calc_idf(docs)
idf_df = pd.DataFrame(list(idf_dict.items()), columns=["Term", "IDF"])
print("\nInverse Document Frequency:")
print(idf_df)

# Creating a TF-IDF model
tfidf_vectorizer = TfidfVectorizer()
tfidf_matrix = round(tfidf_vectorizer.fit_transform(docs), 2)
tfidf_tokens = tfidf_vectorizer.get_feature_names_out()
tfidf_df = pd.DataFrame(data=tfidf_matrix.toarray(), columns=tfidf_tokens)
print("\nTF-IDF Matrix:")
print(tfidf_df)
```

```python
# Calculating Cosine Similarity
cosine_sim_matrix = cosine_similarity(tfidf_matrix)
cosine_sim_df = pd.DataFrame(data=cosine_sim_matrix)
print("\nCosine Similarity Matrix:")
print(cosine_sim_df)
```

```python
# Calculating Manhattan Distance
manhattan_s1_s2 = round(1 / cityblock(tfidf_df.iloc[0], tfidf_df.iloc[1]), 2)
manhattan_s1_s3 = round(1 / cityblock(tfidf_df.iloc[0], tfidf_df.iloc[2]), 2)
manhattan_s2_s3 = round(1 / cityblock(tfidf_df.iloc[1], tfidf_df.iloc[2]), 2)
print("\nManhattan Distance:")
print("S1 and S2:", manhattan_s1_s2)
print("S1 and S3:", manhattan_s1_s3)
print("S2 and S3:", manhattan_s2_s3)
```

```python

# Calculating Euclidean Distance
euclidean_s1_s2 = round(math.dist(tfidf_df.iloc[0], tfidf_df.iloc[1]), 2)
euclidean_s2_s3 = round(math.dist(tfidf_df.iloc[1], tfidf_df.iloc[2]), 2)
euclidean_s1_s3 = round(math.dist(tfidf_df.iloc[0], tfidf_df.iloc[2]), 2)
print("\nEuclidean Distance:")
print("S1 and S2:", euclidean_s1_s2)
print("S2 and S3:", euclidean_s2_s3)
print("S1 and S3:", euclidean_s1_s3)
```
